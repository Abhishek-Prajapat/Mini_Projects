## Loading the Libraries and Data

```python
import pandas as pd
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import plotly_express as px
import seaborn as sns
```

```python
attribute_df      = pd.read_csv('../input/list_attr_celeba.csv')
bbox_df           = pd.read_csv('../input/list_bbox_celeba.csv')
landmark_df       = pd.read_csv('../input/list_landmarks_align_celeba.csv')
eval_partition_df = pd.read_csv('../input/list_eval_partition.csv')
```

### Information corresponding each data-frame

   **1) attribute_df** - Attribute labels for each image. There are 40 attributes. "1" represents positive while "-1" represents negative.
    
   **2) bbox_df** - Bounding box information for each image. "x1" and "y1" represent the upper left point coordinate of bounding box. "width" and "height" represent the width and height of bounding box.
    
   **3) landmark_df** - Image landmarks and their respective coordinates. There are 5 landmarks: left eye, right eye, nose, left mouth, right mouth.
    
   **4) eval_partition_df** - Recommended partitioning of images into training, validation, testing sets. Images 1-162770 are training, 162771-182637 are validation, 182638-202599 are testing.

## Comparing bbox_df with haar_cascade

To make the required model for the real world we will have to make sure that the images on which we train our model can be easily obtained by some algorithm to feed to the model else the model will be useless. For our take we will choose the haar_cascade algorithm and compare it with the current bbox cause if they give same results then we don't need to apply haar_cascade on each image and hence can save computational cost. 

```python
bbox_df.head(2)
```

```python
len(bbox_df)
```

```python
def compare(image_ids, image_dir, model_path):
    
    faceDetector = cv2.CascadeClassifier(model_path)
    n = len(image_ids)
    
    plt.figure(figsize=(15,10))
    pos = 1
    
    for idx, image_id in enumerate(image_ids):
    
        org_image = plt.imread(image_dir + os.listdir(image_dir)[image_id])
        bbox_image = np.copy(org_image)
        haar_image = np.copy(org_image)
        
        y = bbox_df['y_1'][image_id+1]
        h = bbox_df['height'][image_id+1]
        x = bbox_df['x_1'][image_id+1]
        w = bbox_df['width'][image_id+1]
        
        print(x,y,w,h, image_dir + os.listdir(image_dir)[image_id], image_id)

        bbox_image = cv2.rectangle(bbox_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

        faceRegions = faceDetector.detectMultiScale(haar_image, scaleFactor=1.06, minNeighbors=6, minSize=(32, 32), flags=cv2.CASCADE_SCALE_IMAGE)
        
        haar_crop = np.copy(org_image)
        
        for (x, y, w, h) in faceRegions:
            haar_crop = haar_crop[y:y+h, x:x+w]
            cv2.rectangle(haar_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        
        plt.subplot(n,4, pos)
        plt.title('Original')
        plt.axis('off')
        plt.imshow(org_image)
        pos += 1
        plt.subplot(n,4, pos)
        plt.title('Given bbox')
        plt.axis('off')
        plt.imshow(bbox_image)
        pos += 1
        plt.subplot(n,4,pos)
        plt.title('result of haar cascade')
        plt.axis('off')
        plt.imshow(haar_image)
        pos += 1
        plt.subplot(n,4,pos)
        plt.title('Haar Cascade croped')
        plt.axis('off')
        plt.imshow(haar_crop)
        pos += 1
```

```python
image_ids = np.random.randint(0, 20000, size=3)
compare(image_ids, image_dir, '../input/haarcascade_frontalface_default.xml')
```

### Observations

For some reason the bbox given in the dataset is not correct and hence not given accurate results. We will therefore create our training dataset by cropping the haar_cascade of the original image and then use it infer the Heavy_Makeup and Wearning_Lipstick. <br>
